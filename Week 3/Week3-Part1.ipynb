## Segmenting and Clustering Neighborhoods in Toronto (Part 1)

### **Importing libraries and packages**

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

### **Webscraping from Wikipedia**

```python
toronto_url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
data  = requests.get(toronto_url).text 
soup = BeautifulSoup(data,"html5lib")
```

### **Appending data into a list**

```python
table_contents=[]
tables=soup.findAll('table')
table = tables[0]
# print(table)
for row in table.findAll('td'):
    cell = {}
    if row.span.text=='Not assigned':
        pass
    else:
        cell['PostalCode'] = row.p.text[:3]
        cell['Borough'] = (row.span.text).split('(')[0]
        cell['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)
```

### **Data Wrangling and Cleaning**

```python
places = pd.DataFrame(table_contents)
places.groupby(["PostalCode", "Borough"]).apply(lambda x: list(np.unique(x)))
places = places.dropna()
places = places.reset_index(drop=True)
```

```python
for index, row in places.iterrows():
    if row["Neighborhood"] == "Not assigned":
        row["Neighborhood"] = row["Borough"]
print("The shape of the dataframe: ", places.shape)
places.to_csv('Toronto.csv')
places.head()
```
