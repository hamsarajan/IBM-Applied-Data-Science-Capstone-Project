## Segmenting and Clustering Neighborhoods in Toronto (Part 2)

### **Importing libraries and packages** 

```python
import pandas as pd
import numpy as np
```

### **Importing data from saved csv files**

```python
places = pd.read_csv("Toronto.csv", index_col=0)
coordinates = pd.read_csv("Geospatial_Coordinates.csv")
```

```python
places.head()
```

```python
coordinates.head()
coordinates.columns
```

### **Extracting and appending latitude and longitude coordinates to the existing dataframe, places**

```python
latitude, longitude = [], []
for index_place, row_place in places.iterrows():
    for index_coord, row_coord in coordinates.iterrows():
        if row_place["PostalCode"] == row_coord["Postal Code"]:
            latitude.append(row_coord["Latitude"])
            longitude.append(row_coord['Longitude'])
```

```python
places["Latitude"] = latitude
places['Longitude'] = longitude
places.to_csv('Toronto_with_coord.csv')
places.head()
```
